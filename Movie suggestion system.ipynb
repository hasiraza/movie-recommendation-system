```python
import pandas as pd
```

```python
movie=pd.read_csv("top10K-TMDB-movies.csv")
```

```python
movie.sample(5)
```

```python
filtered_data=movie[['id','title','genre','overview']]
```

```python
filtered_data['tag']=    filtered_data['overview']+filtered_data['genre']
```

```python
filtered_data
```

```python
new_data=filtered_data[['id','title','tag']]
```

```python
new_data.isnull().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000, stop_words='english')
vector=cv.fit_transform(new_data['tag'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarty=cosine_similarity(vector)
```

```python
similarty
```

```python
distance=sorted(list(enumerate(similarty[2])), reverse=True, key=lambda vector: vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance=sorted(list(enumerate(similarty[index])), reverse=True, key=lambda vector: vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Weathering with You")
```
